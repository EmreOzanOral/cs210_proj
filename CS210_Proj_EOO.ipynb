```python
import pandas as pd
import json

# Function to read and process a Spotify data file
def read_spotify_data(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        data = json.load(file)

    # Create a DataFrame from the list of song entries
    df = pd.DataFrame(data)
    
    # Convert 'ts' column to datetime format
    df['ts'] = pd.to_datetime(df['ts'])
    
    return df

# List of file paths for files 1.json to 10.json
file_paths = ['1.json', '2.json', '3.json', '4.json', '5.json', '6.json', '7.json', '8.json', '9.json', '10.json']

# List to store DataFrames for each file
dataframes = []

# Loop through each file and create a DataFrame
for file_path in file_paths:
    df = read_spotify_data(file_path)
    dataframes.append(df)

# Concatenate DataFrames into a single DataFrame
final_dataframe = pd.concat(dataframes, ignore_index=True)

# Display the final DataFrame

final_dataframe


```

```python
from scipy.stats import ttest_ind

musical_value_columns = ['Loud', 'Happy', 'Energy', 'Dance', 'BPM']

for column in musical_value_columns:
    threshold = most_df[column].median()
    
    # Check if there is more than one unique value
    if most_df[column].nunique() > 1:  
        high_values = most_df[most_df[column] > threshold]
        low_values = most_df[most_df[column] <= threshold]

        # Independent t-test
        t_stat, p_value = ttest_ind(high_values['Popularity'], low_values['Popularity'])
        
        # Print results
        print(f'Test for {column}: T-statistic = {t_stat}, p-value = {p_value}')

        # Check for significance at a 0.05 level (adjust as needed)
        if p_value < 0.05:
            print(f"The difference in Popularity for {column} is statistically significant.")
        else:
            print(f"The difference in Popularity for {column} is not statistically significant.")
```

```python
from scipy.stats import f_oneway

musical_value_columns = ['Loud', 'Happy', 'Energy', 'Dance', 'BPM']

for column in musical_value_columns:
    threshold = most_df[column].median()
    
    # Check if there is more than one unique value
    if most_df[column].nunique() > 1:
        high_values = most_df[most_df[column] > threshold]
        low_values = most_df[most_df[column] <= threshold]

        # ANOVA test
        f_stat, p_value = f_oneway(high_values['Popularity'], low_values['Popularity'])
        
        # Print results
        print(f'ANOVA for {column}: F-statistic = {f_stat}, p-value = {p_value}')

        # Check for significance at a 0.05 level (adjust as needed)
        if p_value < 0.05:
            print(f"There is a significant difference in Popularity across {column} groups.")
        else:
            print(f"There is no significant difference in Popularity across {column} groups.")
```

```python
from scipy.stats import spearmanr

musical_value_columns = ['Loud', 'Happy', 'Energy', 'Dance', 'BPM']

for column in musical_value_columns:
    threshold = most_df[column].median()
    
    # Check if there is more than one unique value
    if most_df[column].nunique() > 1:  
        correlation, p_value = spearmanr(most_df[column], most_df['Popularity'])
        
        # Print results
        print(f'Correlation for {column}: {correlation:.4f}, p-value = {p_value:.4f}')

        # Check for significance at a 0.05 level (adjust as needed)
        if p_value < 0.05:
            print(f"The correlation between Popularity and {column} is statistically significant.")
        else:
            print(f"The correlation between Popularity and {column} is not statistically significant.")
```

```python
from scipy.stats import pearsonr

musical_value_columns = ['Loud', 'Happy', 'Energy', 'Dance', 'BPM']

for column in musical_value_columns:
    threshold = most_df[column].median()
    
    # Check if there is more than one unique value
    if most_df[column].nunique() > 1:  
        correlation, p_value = pearsonr(most_df[column], most_df['Popularity'])
        
        # Print results
        print(f'Correlation for {column}: {correlation:.4f}, p-value = {p_value:.4f}')

        # Check for significance at a 0.05 level (adjust as needed)
        if p_value < 0.05:
            print(f"The correlation between Popularity and {column} is statistically significant.")
        else:
            print(f"The correlation between Popularity and {column} is not statistically significant.")
```

```python
import statsmodels.api as sm

musical_value_columns = ['Loud', 'Happy', 'Energy', 'Dance', 'BPM']
alpha = 0.05  # Adjust as needed

for column in musical_value_columns:
    X = most_df[[column]]
    y = most_df['Popularity']

    X = sm.add_constant(X)  # Add a constant term to the predictor

    model = sm.OLS(y, X).fit()

    # Print results
    print(f'Regression for {column}: Coefficient = {model.params[1]:.4f}, Intercept = {model.params[0]:.4f}')
    print(f'R-squared value: {model.rsquared:.4f}')
    print(f'P-value for {column}: {model.pvalues[1]:.4f}')

    # Check for significance
    if model.pvalues[1] < alpha:
        print(f"The coefficient for {column} is statistically significant.")
    else:
        print(f"The coefficient for {column} is not statistically significant.")
```

```python
import statsmodels.api as sm

X = most_df[['Loud', 'Happy', 'Energy', 'Dance', 'BPM']]
y = most_df['Popularity']

X = sm.add_constant(X)  # Add a constant term to the predictors

model = sm.OLS(y, X).fit()

# Print results
print('Multiple Regression:')
for i, column in enumerate(musical_value_columns):
    print(f'Coefficient for {column}: {model.params[i+1]:.4f}')

print(f'Intercept: {model.params[0]:.4f}')
print(f'R-squared value: {model.rsquared:.4f}')
print('P-values for coefficients:')
for i, column in enumerate(musical_value_columns):
    print(f'{column}: {model.pvalues[i+1]:.4f}')

# Check for significance
significant_coeffs = [column for i, column in enumerate(musical_value_columns) if model.pvalues[i+1] < alpha]
if significant_coeffs:
    print(f"The coefficients for {', '.join(significant_coeffs)} are statistically significant.")
else:
    print("No statistically significant coefficients.")
```
